# Homicidal-Happiness
This project takes two datasets on international happiness levels and homicide rates, combines them, and attempts to figure out what the happiest countries are, what the most dangerous countries are, and is there a correlation between happiness levels and homicide rates.

```python
# import depedencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# load in dataset 1
happiness_df = pd.read_csv('happiness.csv')
happiness_df.tail()
```

```python
# change happiness_df column names
happiness_df = happiness_df.rename(columns={'Countries':'Country','Happiness index, 2022':'Happiness Index','Available data':'Year'})
happiness_df
```

```python
# reorder happiness dataset
happiness_df = happiness_df[['Global rank','Country','Happiness Index','Year']]
happiness_df.tail()
```

```python
# count NaN values
NaN_count = happiness_df.isna().sum()
NaN_count
```

```python
# load in dataset 2
homicide_df = pd.read_csv('homicide.csv')
homicide_df.tail()
```

```python
# rename homicide_df columns
homicide_df = homicide_df.rename(columns={'Location':'Country','Rate':'Homicide Rate (%)','Count':'Homicide Count'})
homicide_df.tail()
```

```python
# sort values by Homicide Rate Percentage and reset index
homicide_df = homicide_df.sort_values('Homicide Rate (%)', ascending=False)
homicide_df = homicide_df.reset_index(drop=True)
homicide_df.head(12)
```

```python
# check for NaN values
NaN_count = homicide_df.isna().sum()
NaN_count
```

```python
# Merge datasets
hh_df = pd.merge(happiness_df, homicide_df, on='Country', how='inner')
hh_df.head(12)
```

```python
NaN_count = hh_df.isna().sum()
NaN_count
```

```python
# cleaned data retention rate analysis
max_record_count = len(homicide_df)
min_record_count = len(happiness_df)
clean_record_count = len(hh_df)
record_retention = clean_record_count/max_record_count*100
record2_retention = clean_record_count/min_record_count*100
record_retention = round(record_retention, 2)
record2_retention = round(record2_retention, 2)
print(f'Original Happiness Records: {min_record_count} records')
print(f'Original Homicide Records: {max_record_count} records')
print(f'Clean Record Count: {clean_record_count} records')
print(f'Cleaned data record retention: {record_retention}% based on Homicide.csv')
print(f'Cleaned data record retention: {record2_retention}% based on Happiness.csv')
```

```python
# export cleaned and merged dataframe as csv
hh_df.to_csv('homicidal-happiness.csv')
```

```python
# plot country happiness/homicide barchart
# create data
x = hh_df["Country"].head(12)
y1 = hh_df["Happiness Index"].head(12)
y2 = hh_df["Homicide Rate (%)"].head(12)
legends = ['Happiness Index','Homicide Rate']
# figure sizing, labels, and legend
plt.figure(figsize=(10,6))
width = .5
plt.xticks(rotation=90)
plt.title("Top 12 Countries by Happiness Index and Their Corresponding Homicide Rates")
plt.ylabel("Happiness Index and Suicide Rates")
plt.xlabel("Countries")
plt.legend(["Happiness Index", "Homicide Rate Percentage"], title ="Attributes")
# plot data in grouped manner of bar type
plt.bar(x, y1, width, color = "r")
plt.bar(x, y2, width, color = "b")
plt.show()
```

```python
# plot top 12 most dangerous countries by homicide rate barchart
name = homicide_df['Country'].head(12)
happiness = homicide_df['Homicide Rate (%)'].head(12)
fig = plt.figure(figsize =(10, 7))
width = 0.8
plt.xticks(rotation=90)
plt.title("Top 12 Most Dangerous Countries by Homicide Rate Percentage")
plt.ylabel("Homicide Rate Percentage")
plt.xlabel("Countries")
plt.bar(name,happiness, width)
plt.show()
```

```python
# plot top 12 least dangerous countries by homicide rate barchart
homicide_df = homicide_df.sort_values('Homicide Rate (%)', ascending=True)
name = homicide_df['Country'].head(12)
happiness = homicide_df['Homicide Rate (%)'].head(12)
fig = plt.figure(figsize =(10, 7))
width = 0.8
plt.xticks(rotation=90)
plt.title("Top 12 Safest Countries by Homicide Rate Percentage")
plt.ylabel("Homicide Rate Percentage")
plt.xlabel("Countries")
plt.bar(name,happiness, width)
plt.show()
```
